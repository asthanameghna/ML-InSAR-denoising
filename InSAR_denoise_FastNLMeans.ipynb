# Denoising using Fast NL Means Algorithm

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
plt.style.use('seaborn')
```

```python
image = cv2.imread('/content/drive/MyDrive/filt_topophase.unw.geo.png')
```

```python
dst = cv2.fastNlMeansDenoisingColored(image, None, 11, 6, 7, 21)
```

```python
row, col = 1, 2
fig, axs = plt.subplots(row, col, figsize=(15, 10))
fig.tight_layout()
axs[0].imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
axs[0].set_title('Kilauea_Hawaii-eruption-EQ')
axs[1].imshow(cv2.cvtColor(dst, cv2.COLOR_BGR2RGB))
axs[1].set_title('Fast Means Denoising Kilauea_Hawaii-eruption-EQ')
plt.show()
```

```python
# Changed parameters in dst2

dst2 = cv2.fastNlMeansDenoisingColored(image,None,10,10,7,21)
row, col = 1, 2
fig, axs = plt.subplots(row, col, figsize=(15, 10))
fig.tight_layout()
axs[0].imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
axs[0].set_title('Kilauea_Hawaii-eruption-EQ')
axs[1].imshow(cv2.cvtColor(dst2, cv2.COLOR_BGR2RGB))
axs[1].set_title('Fast Means Denoising Kilauea_Hawaii-eruption-EQ')
plt.show()
```
